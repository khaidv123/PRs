```python
pulls = repo.get_pulls(state='all')
```

```python
data= []

for pr in pulls:
  data.append({"ID": pr.number})

df= pd.DataFrame(data)
df
```

```python
data1= []

for pr in pulls:
  data1.append({"Created": pr.created_at})

df1= pd.DataFrame(data1)
df1
```

```python
data2= []

for pr in pulls:
  closed_at = pr.closed_at
  data2.append({"Closed": closed_at})

df2= pd.DataFrame(data2)
df2
```

```python
data3= []

for pr in pulls:
  created_at = pr.created_at
  closed_at = pr.closed_at
  time_to_close_hours = None
  if closed_at:
      time_to_close_hours = (closed_at - created_at).total_seconds() / 3600.0
  data3.append({"Time To Close (hours)": time_to_close_hours})

df3= pd.DataFrame(data3)
df3
```

```python
data4= []

for pr in pulls:
  changed_files_count = pr.changed_files
  data4.append({"Number of Changed Files": changed_files_count})

df4= pd.DataFrame(data4)
df4
```

```python
data5= []

for pr in pulls:
  data5.append({"Title": pr.title})

df5= pd.DataFrame(data5)
df5
```

```python
data6= []

for pr in pulls:
  labels = [label.name for label in pr.labels]
  data6.append({"Labels": labels})

df6= pd.DataFrame(data6)
df6
```

```python
data7= []

for pr in pulls:
  labels = [label.name for label in pr.labels]
  labels_count = len(labels)
  data7.append({"Labels#": labels_count})

df7= pd.DataFrame(data7)
df7
```

```python
data8= []

for pr in pulls:
  is_merged = pr.merged
  data8.append({"Is Merged": is_merged})

df8= pd.DataFrame(data8)
df8
```

```python
data9= []

for pr in pulls:
   data9.append({'Time to Merge (hours)': (pr.merged_at - pr.created_at).total_seconds() / 3600 if pr.merged_at else None
    })

df9= pd.DataFrame(data9)
df9
```

```python
data10= []

for pr in pulls:
  base_ref = pr.base.ref
  description_exists = bool(pr.body)
  open_or_closed = pr.state
  data10.append({"Base": base_ref,
        "Descriptions": description_exists,
        "Open/Closed": open_or_closed})

df10= pd.DataFrame(data10)
df10
```

```python
data11= []

for pr in pulls:
  data11.append({'Author Name': pr.user.name if pr.user else None})

df11= pd.DataFrame(data11)
df11
```

```python
data12= []

for pr in pulls:
  data12.append({"Updated at": pr.updated_at})

df12= pd.DataFrame(data12)
df12
```

```python
data13= []

for pr in pulls:
  data13.append({"Commits": pr.commits})

df13= pd.DataFrame(data13)
df13
```

```python
commits = repo.get_commits()

# Get the commit messages:
commit_messages = [commit.commit.message for commit in commits]

# Create a pandas dataframe from the commit messages:
df_commit = pd.DataFrame(commit_messages, columns=["Commit Message"])
df_commit.to_csv("CommitMessage.csv")
```

```python
pr = pulls
#commits = pr.get_commits()
commits = repo.get_commits()

file_changes = {}
for commit in commits:
    files = commit.files
    for file in files:
        filename = file.filename
        additions = file.additions
        deletions = file.deletions
        if filename not in file_changes:
            file_changes[filename] = additions + deletions
        else:
            file_changes[filename] += additions + deletions

df_ = pd.DataFrame.from_dict(file_changes, orient='index', columns=['Total Change'])
df_.to_csv("FiledChanged.csv")
```
